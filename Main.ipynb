```python
# this is responsible for descovring all the leds on the network


from sdk_python.API import API
api = API()

devices = api.get_online_devices()
```

```python
selected_device = devices[0]
selected_device.__dict__
```

```python
message = api.message_builder.build_setting_command(0.1)
api.network.send(selected_device.address, message.to_bytes())
```

```python
message = api.message_builder.build_rgb_stream_command(0.5, 0, [255, 196, 0] * 150)
api.network.send(selected_device.address, message.to_bytes())
```

```python
message = api.message_builder.build_rgb_stream_command(0.5, 0, [255, 0, 196] * 150)
api.network.send(selected_device.address, message.to_bytes())
```

```python
message = api.message_builder.build_rgb_stream_command(0.5, 0, [255, 0, 196] * 75 + [255, 196, 0] * 75)
api.network.send(selected_device.address, message.to_bytes())
```
